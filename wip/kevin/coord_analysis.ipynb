```python
xq95 = coord["x_d"].quantile(0.95)
xq05 = coord["x_d"].quantile(0.05)
yq95 = coord["y_d"].quantile(0.95)
yq05 = coord["y_d"].quantile(0.05)
tq95 = coord["theta_d"].quantile(0.95)
tq05 = coord["theta_d"].quantile(0.05)


x_outliers = (coord["x_d"] > xq95) | (coord["x_d"] < xq05)
y_ouliers = (coord["y_d"] > yq95) | (coord["y_d"] < yq05)
theta_outliers = (coord["theta_d"] > tq95) | (coord["theta_d"] < tq05)

outliers = x_outliers | y_ouliers | theta_outliers

problems = coord[outliers].drop(["x_prev", "y_prev", "x_cur", "y_cur", "theta_prev", "theta_cur"], axis=1)
problems
```

```python
pb_frames_x = (coord[(coord["x_d"] > xq95) | (coord["x_d"] < xq05)]["frame"]).to_frame()
pb_frames_y = (coord[(coord["y_d"] > yq95) | (coord["y_d"] < yq05)]["frame"]).to_frame()
pb_frames_theta = (coord[(coord["theta_d"] > tq95) | (coord["theta_d"] < tq05)]["frame"]).to_frame()
array_pbs = [pb_frames_x, pb_frames_y, pb_frames_theta]
```

```python
#pb_frames = pb_frames_x.merge(pb_frames_y, how='outer')
pb_frames = pd.concat(array_pbs, axis=1, keys=['xframe', 'yframe', 'thetaframe'])

pb_frames
```
